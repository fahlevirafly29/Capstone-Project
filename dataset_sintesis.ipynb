```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv('dataset_after_wrangling.csv')
```

Ringkasan Statistik / Angka Kasar

```python
print("Ringkasan Angka (Umur, Pengalaman, Skor)")
display(df.describe())
```

```python
print("\n Ringkasan Teks (Pekerjaan, Pendidikan, Jurusan)")
display(df.describe(include='O'))
```

Visualisasi Distribusi Profesi

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y='job', order=df['job'].value_counts().index, palette='magma')
plt.title('Jumlah Pelamar per Profesi')
plt.xlabel('Jumlah Orang')
plt.ylabel('Nama Profesi')
plt.show()
```

Persebaran Umur

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['age'], bins=20, kde=True, color='teal')
plt.title('Persebaran Umur Pelamar')
plt.xlabel('Umur')
plt.ylabel('Jumlah Pelamar')
plt.show()
```
