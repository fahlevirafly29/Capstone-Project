```python
import pandas as pd
import ast
```

Gathering Data

```python
df = pd.read_csv('synthetic_data_4.csv')
```

Cleaning Data

```python
df = df.drop_duplicates()
df = df.dropna(subset=['job'])
```

```python
print("Shape setelah drop duplikat:", df.shape)
```

```python
df['age'] = df['age'].clip(lower=18, upper=60)
df['skill_score'] = df['skill_score'].clip(lower=0, upper=100)
```

```python
def parse_list(x):
    if pd.isna(x): return []
    try: return ast.literal_eval(x)
    except: return []

for col in ['skills', 'certifications', 'projects']:
    df[col] = df[col].apply(parse_list)
```

```python
print("Tipe data skills sekarang:", type(df['skills'].iloc[0]))
```

```python
df['age'] = df['age'].fillna(df['age'].median())
df['years_experience'] = df['years_experience'].fillna(df['years_experience'].median())
df['skill_score'] = df['skill_score'].fillna(df['skill_score'].median())

for col in ['education_level', 'major', 'summary', 'experience_desc']:
    df[col] = df[col].fillna('unknown')
```

```python
print("Sisa missing value per kolom:\n", df.isna().sum())
```

```python
for col in ['job', 'education_level', 'major']:
    df[col] = df[col].astype(str).str.lower().str.strip()

print("Sisa missing value:\n", df.isna().sum())
print("\nShape setelah cleaning:", df.shape)
```

```python
df = df[~df.astype(str).duplicated()]
```

```python
print("hasil cek akhir")
print("Shape final:", df.shape)
print("Total Duplikat:", df.astype(str).duplicated().sum())
print("Total Missing Value:", df.isna().sum().sum())
```

```python
print("1. Cek Info dan Tipe Data")
df.info()

print("\n2. Cek Sisa Data Kosong dan Duplikat")
print("Sisa data kosong:", df.isna().sum().sum())
print("Sisa data duplikat:", df.astype(str).duplicated().sum())

print("\n3. Cek Rentang Nilai Umur dan Skor")
print("Umur terkecil dan terbesar:", df['age'].min(), "-", df['age'].max())
print("Skor terkecil dan terbesar:", df['skill_score'].min(), "-", df['skill_score'].max())

print("\n4. Lihat 3 Baris Teratas")
display(df.head(3))
```
